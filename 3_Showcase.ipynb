Начинаем собирать витрину по частям. Сначал подтягиваем легкодоступные данные из основного датафрейма

```python
cols_2 = ['Revenue','Marge', 'Телефон_new','КоличествоПроданоКлиенту']

showcase = df[cols_2].groupby(['Телефон_new']).agg(sum)
```

```python
showcase
```

```python
showcase = showcase.reset_index()
showcase
```

Подтягиваем уникальные чеки

```python
showcase1 = df[['Телефон_new','НомерЗаказаНаСайте']].groupby('Телефон_new').agg(np.unique) #количество уник чеков
list_bills = []
for i in showcase1['НомерЗаказаНаСайте']:
  if (type(i) == str):
    d = []
    d.append(i)
    list_bills.append(len(d))
  else:
      list_bills.append(len(i))

print(list_bills)
```

```python
len(list_bills)
```

```python
showcase["КоличествоЧеков"] = list_bills
showcase
```

Высчитываем дополнительные параметры

```python
showcase['СреднийЧек'] = showcase['Revenue']/showcase['КоличествоЧеков']
showcase['СреднийЧислоТоваровЧека'] = showcase['КоличествоПроданоКлиенту']/showcase['КоличествоЧеков']
showcase['СредняяМаржа'] = showcase['Marge']/showcase['КоличествоЧеков']
showcase
```

Займемся частью данных, отвечаюзих за категории. Посчитаем кол-во покупок для каждого клиента относительно группы3. считать будем сначалу сумму товаров в каждой категории, затем сколько это от общего числа покупок клиента




```python
category = df[['Телефон_new','Группа3','КоличествоПроданоКлиенту']].pivot_table('КоличествоПроданоКлиенту',index ='Телефон_new', columns = 'Группа3', aggfunc=np.sum)

category = category.reset_index()
category = category.fillna(0)
category
```

Склеиваем наши датафреймы

```python
showcase = showcase.merge(category, how='right', on='Телефон_new')
showcase
```

Тоже самое делаем для формы оплаты и местоположения

```python
dummies = pd.get_dummies(df['ФормаОплаты'])
df = pd.concat([df, dummies], axis=1)
df.drop('ФормаОплаты', axis=1, inplace=True)

df.head()
```

```python
payment = df[['Телефон_new','Безналичная','Наличная']].groupby('Телефон_new').agg(sum) 
payment.reset_index()
payment['sum_operations'] = payment['Безналичная']+payment['Наличная']
payment
```

```python
showcase = showcase.merge(payment, how='right', on='Телефон_new')
showcase['Безналичная'] = showcase['Безналичная']/showcase['sum_operations']
showcase['Наличная'] = showcase['Наличная']/showcase['sum_operations']
```

```python
dummies = pd.get_dummies(df['Гео'])
df = pd.concat([df, dummies], axis=1)
df.drop('Гео', axis=1, inplace=True)
df.head()
```

```python
geo = df[['Телефон_new','МО','Москва','Регионы']].groupby('Телефон_new').agg(sum) 
geo.reset_index()
```

```python
showcase = showcase.merge(geo, how='right', on='Телефон_new')
showcase['МО'] = showcase['МО']/showcase['sum_operations']
showcase['Москва'] = showcase['Москва']/showcase['sum_operations']
showcase['Регионы'] = showcase['Регионы']/showcase['sum_operations']
```

```python
showcase = showcase.drop(['sum_operations'], axis=1)
showcase
```
